```python
import numpy as np
def PCA(X,num_components):
    #step 1
    X_meaned=X-np.mean(X,axis=0)
    #step 2
    cov_mat=np.cov(X_meaned,rowvar=False)
    #step 3
    eigen_values,eigen_vectors=np.linalg.eigh(cov_mat)
    # step 4
    sorted_index=np.argsort(eigen_values)[::-1]
    sorted_eigenvalue=eigen_values[sorted_index]
    sorted_eigenvectors=eigen_vectors[:,sorted_index]
    #step 5
    eigenvector_subset=sorted_eigenvectors[:,0:num_components]
    #step 6
    X_reduced=np.dot(eigenvector_subset.transpose(),X_meaned.transpose()).transpose()
    return X_reduced
```

```python
import pandas as mypd
mydata=mypd.read_csv("Iris_data.csv")
mydata
```

```python
#prepare the data
X=mydata.iloc[:,0:4]
#prepare the target
target=mydata.iloc[:,4]
#applying it to PCA function
mat_reduced=PCA(X,3)
#creating a pandas df of reduced dataset
principal_df=mypd.DataFrame(mat_reduced,columns=['PC1','PC2','PC3'])
# concat it with target variable to create a complete datset
principal_df=mypd.concat([principal_df,mypd.DataFrame(target)],axis=1)
```
